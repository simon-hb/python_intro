## Regular Expressions

Regular Expressions or "Regex" is a very powerful tool used to match characters to pieces of text. We use Regex to define a set of rules that we want to place on pieces of text for cleaning and validation. The Python ecosystem has a built-in library called `re` that gives us the power of Regular Expressions. 

In addition to matching regular characters to strings there are special characters known as metacharacters. Metacharacters are exceptions to the rule of normal character matching. Instead metacharacters do not match themselves but rather signal that some out-of-the-ordinary meaning should be matched.

### Basic Metacharacters

<img src='https://drive.google.com/uc?view=export&id=1Vixgb6xSF3nYrVDD9etI01RpEOrM0h5B' />

````python
import re 

# find all instances of 'ai' in a string
string = "It's raining in Spain"
x = re.findall('ai', string)
print(x)
````

````python
# Replace a word
sunny = re.sub('raining', 'sunny', string)
print(sunny)
````

````python
# find all phone numbers using metacharacters
document = 'You can contact me via my cell phone: (555)-432-1234 or office: (555)-432-5678'
phone_numbers = re.findall("\(\d\d\d\)-\d\d\d-\d\d\d\d", document)
print(phone_numbers)
````

Python has a special form of string called a raw string. It is best to use the `r` prefix when writing regular expressions to ensure that your expression is treated as a raw string. A raw string in a regular expression will treat backslash characters literally instead of signifying special treatment to the following character. A good example of this is the `\n` characters. Normally `\n` signals is a new line but when prefixing with the r decorator, r'\n' means two characters: a backslash \ followed by n.

## Activity: Regular Expressions

You're given a string that contains a person's contact information. Using the provided Regular Expressions, use Python's `re` library to extract and print the user's email address and phone number. Note that this would be quite difficult to do without regular expressions. It's not necessary to be able to read and understand the regex strings.
- Import the `re` library
- You will need to use the .findall() method for each Regular Expression


````python
contact = """
Name:  Alex Apricot
Email: aapricot@gmail.com
Phone: +1 (123) 456 7890
Addr:  123 Four St.
       Nowhere, ON
       Canada A1A 1A1
"""

# Write Code Here

import re
email_address = re.findall("\w+\:\s+\w+@\w+",contact)  # + must match at least once but will match more
phone_number = re.findall("(\d\s+\(\d{3,3}\)\s+\d{3,3}\s+\d{4,4})",contact) # all this depends on way data is coming in - what does this code assume?
print(email_address)
print(phone_number)          
````

## matpoltlib

Matplotlib is a Python data visualization library built on NumPy arrays that enables us to embed graphs directly into Notebooks. To get started with matplotlib we need to import the library by writing the following:

```py
import matplotlib.pyplot as plt
```

We will be using the following Pandas DataFrame to demonstrate how to create visualizations with matplotlib. It is important to know that the .plot() method in these examples are associated with Pandas while the .plt() is associated with the imported matplotlib library.

Pandas have their own functionality built into their library which leverages matplotlib. This means that we do not need to import matplotlib when using Pandas but do note that .plt() and .plot() are the same function.

````python
import pandas as pd
import numpy as np

df_one = pd.DataFrame({"A": np.random.randint(75,200,50), "B": np.random.randint(0,25,50), "C": np.random.randint(-50,15,50) })
df_one.head(10)
````

````python
df_one.plot()
````

### Plotting charts

Simply calling the .plot() method on a DataFrame will generate a line chart as seen above. If we wanted to plot using the imported matplotlib and a NumPy array, do the following:

````python
import matplotlib.pyplot as plt

numpy_array = df_one.values
plt.plot(numpy_array)
````

By default, using the .plot() method will connect all data points with lines to create a continuous line graph. If we do not specific colors matplotlib will automatically assign default colors to our individual datasets.

### Plot a single column

We can plot a single column from our DataFrame through indexing. Refer to the example below.

````python
df_one['A'].plot()
````

To avoid printing out the matplotlib object `<matplotlib.axes._subplots.AxesSubplot at 0x119d55780>`, place a semi-colon `;` at the end of your plot command. This makes for a cleaner presentation of your chart.

````python
df_one['B'].plot();
````

### Basic Customization: matplotlib & Pandas

Adding some basic customizations to your charts is made easy. Simply pass the plot function varies arguments such as a title, color or legend. In addition, we can add labels to our axes by setting our plot to a variable to a variable and call the .set_xlabel() or .set_ylabel() methods.

````python
line_plot = df_one["C"].plot(title='Column C', color='purple', legend=True);
line_plot.set_xlabel("This is my X label");
line_plot.set_ylabel("This is my Y label");
````

### Types of Charts

There are a wide range of charts available, here are some of the most common charts used:
- Line plot: dataframe.plot(...)
- Histogram: dataframe.plot.hist(...)
- Bar charts: dataframe.plot.bar(...)
- Pie charts: dataframe.plot.pie(...)
- Scatter plot: dataframe.plot.scatter(...)

For the entire list of charts and arguments available for matplotlib with Pandas visit the [Pandas Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html).
